```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
```

```python
spark=SparkSession.builder\
    .master("local[*]")\
    .appName("WordCount")\
    .getOrCreate()
```

```python
sc=spark.sparkContext
```

```python
detik_com_file="rusia-serang-ukraina.html"
```

```python
detikcom_rdd=sc.textFile(detik_com_file)
```

```python
detikcom_rdd.count()
```

```python
import re
TAG_RE = re.compile(r'<[^>]+>')
```

```python
def lower_clean_str(x):
  punc='!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~-'
  lowercased_str = x.lower()
  for ch in punc:
    lowercased_str = lowercased_str.replace(ch, '')
    lowercased_str = TAG_RE.sub('', lowercased_str)
  return lowercased_str
```

```python
detikcom_rdd = detikcom_rdd.map(lower_clean_str)
```

```python
detikcom_rdd.take(40)
```

```python
detikcom_rdd=detikcom_rdd.flatMap(lambda satir: satir.split(" "))
```

```python
detikcom_rdd.take(5)
```

```python
detikcom_rdd = detikcom_rdd.filter(lambda x:x!='')
```

```python
detikcom_rdd.take(4)
```

```python
detikcom_count=detikcom_rdd.map(lambda  word:(word,1))
```

```python
detikcom_count.take(4)
```

```python
detikcom_count_RBK=detikcom_count.reduceByKey(lambda x,y:(x+y)).sortByKey()
```

```python
detikcom_count_RBK.take(40)
```

```python
detikcom_count_RBK.sortByKey(False).take(10)
```
